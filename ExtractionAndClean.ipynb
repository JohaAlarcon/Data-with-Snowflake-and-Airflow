```python
import pandas as pd
import time
import random
import os
from datetime import datetime
import uuid
import re
```

```python
url =['https://www.espn.com.co/futbol/posiciones/_/liga/col.1',
      'https://www.espn.com.co/futbol/posiciones/_/liga/eng.1/liga',
      'https://www.espn.com.co/futbol/posiciones/_/liga/ita.1/liga',
      'https://www.espn.com.co/futbol/posiciones/_/liga/ger.1/liga',
      'https://www.espn.com.co/futbol/posiciones/_/liga/fra.1/liga'
]

ligas = ['Colombia','Inglaterra', 'Italia', 'Alemania','Francia']

df_ligas = {
    'LIGA': ligas,
    'URL': url
}

df_ligas = pd.DataFrame(df_ligas)
df_ligas
```

```python
df = pd.concat([df[0],df[1]], ignore_index = True, axis=1)
df
```

```python
df = df.rename (columns ={0:'Equipo', 1:'J', 2:'G', 3:'E', 4:'P', 5:'GF', 6:'GC', 7:'DIF', 8:'PTS' })

remove_leading_digits_and_uppercase_except_last = lambda x: re.sub(r'^\d+|[A-Z](?=[A-Z])|[A-Z](?=$)','',x)
df['Equipo'] = df['Equipo'].apply(remove_leading_digits_and_uppercase_except_last)

print(df['Equipo'])
```
